### Hashmap Implementation

```python
hash("abcd")
```

```python
class HashMap:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]  # list of buckets

    def _hash(self, key):
        return hash(key) % self.size

    def put(self, key, value):
        index = self._hash(key)
        # check if key already exists â†’ update
        for pair in self.table[index]:
            if pair[0] == key:
                pair[1] = value
                return
        # otherwise, insert new pair
        self.table[index].append([key, value])

    def get(self, key):
        index = self._hash(key)
        for pair in self.table[index]:
            if pair[0] == key:
                return pair[1]
        return None  # key not found

    def remove(self, key):
        index = self._hash(key)
        for i, pair in enumerate(self.table[index]):
            if pair[0] == key:
                del self.table[index][i]
                return

    def display(self):
        for i, bucket in enumerate(self.table):
            print(i, ":", bucket)
```

```python
hm = HashMap()

hm.put("Raj", 20)
hm.put("Anu", 25)
hm.put("Tom", 30)

hm.display()
# Example Output:
# 0 : []
# 1 : [['Anu', 25]]
# 2 : [['Raj', 20], ['Tom', 30]]  # collision handled by chaining
# ...

print(hm.get("Raj"))   # 20
print(hm.get("Anu"))   # 25

hm.remove("Tom")
hm.display()
```
